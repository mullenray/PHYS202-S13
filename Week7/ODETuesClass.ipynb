###Approximate Solution

```python
g = 9.8 #m/s^2
l = 0.5 #m
th0 = 20.0 * pi / 180 #20 degrees expressed in radians

t = arange(0.0, 10.0 + 0.01, 0.01)
theta_app = lambda t: th0 * cos(sqrt(g / l) * t)
    
plot(t, theta_app(t))
show()
```

###Numerical Solution

```python
def deriv(theta_current, t):
    """Return derivatives of the array theta"""
    g = 9.8 #m/s^2
    l = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l) * sin(theta)
    return array( [dtheta, domega] )
```

```python
from scipy.integrate import odeint
```

```python
t = arange(0.0, 10.0 + 0.01, 0.01)

omega0 = 0.0 #Starting from rest, omega = 0
theta_init = array( [th0, omega0] )

theta = odeint(deriv, theta_init, t)
```

```python
theta.shape
```

```python
plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
show()
```

```python
th0 = 50.0 * pi / 180 #50 degrees in radians

theta_init = array( [th0, omega0] ) #Initial values
theta = odeint(deriv, theta_init, t)

plot(t, theta_app(t), 'b-', t, theta[:,0], 'r--')
show()
```

Example 1: Analytic Inverse CDF Transform
=

```python
x = arange(0.0, pi + 0.001, 0.01)
f = lambda x: sin(x) / 2.0
F = lambda x: (cos(0) - cos(x)) / 2.0

plot(x, f(x), label = 'True distribution', lw = 3)
plot(x, F(x), label = 'Cumulative function', lw = 3)
     
xlim(0,3.5)
ylim(0,1.1)

legend(loc = 0)
show()
```

```python
#Number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z : arccos(cos(0) - 2 * z)
    
#sampled x values
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples, 50, normed = True, label = 'Sampled distribution')
plot(x, f(x), label = 'True distribution', lw = 3, color = 'red')

xlim(0, pi)
ylim(0, 0.7)

legend(loc = 0)
show()
```

Exercise 1. Exponential function - analytic
-
###(a)

$F(x) = \int^{x}_{0}\lambda e^{-\lambda t}dt = 1 - e^{-\lambda x}$

$F^{-1}(z) = \frac{- ln(1 - z)}{\lambda}$

###(b)

```python
x = arange(0.0, 5.001, 0.01)
lam = 1.
f = lambda x: lam * exp(-lam * x)
F = lambda x: 1 - exp(-lam * x)

plot(x, f(x), label = 'True distribution', lw = 3)
plot(x, F(x), label = 'Cumulative function', lw = 3)
     
xlim(0, 5.1)

legend(loc = 0)
show()
```

###(c)

```python
N = 10000

z = numpy.random.random_sample(N);

Finv = lambda z: (-log(1-z))/lam
    
xsamples = Finv(z);

hist(xsamples, 50, normed = True, label = 'Sampled distribution')
plot(x, f(x), label = 'True distribution', lw = 3, color = 'red')

xlim(0, 5.01)

legend(loc = 0)
show()
```

```python
x = arange(0.0, 5.001, 0.01)
lam = 1.
f = lambda x: lam * exp(-lam * x)
    
disnum = numpy.array(f(x))

cdfnum = cumsum(disnum)/sum(disnum)

N = 10000
z = numpy.random.random_sample(N)
indval = searchsorted(cdfnum, z)
print indval
samx = numpy.array(x[indval])
```

###(c)

```python
def distribute1D(x, prob, N):
    """Takes any numerical distribution prob, on the interval defined by the array x and returns an array of N sampled values that are statistically the same as the input data."""
    cdfnum = cumsum(prob)/sum(prob)
    z = numpy.random.random_sample(N)
    indval = searchsorted(cdfnum, z)
    xsamples = numpy.array(x[indval])
    return xsamples
```

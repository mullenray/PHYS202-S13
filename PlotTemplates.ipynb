###Line Plots

```python
y = arange(-5,5, 0.01) #create a range in the x-axis over which to plot data
E = cosh(y)
p = sinh(y)
```

```python
plot(y, p, color = "green", label = "Label1")  #Plotting the "p" function in a green color
plot(y, E, color = "red", label = "Label2")  #Plotting the "E" function in a red color
xlabel("X-axis label", fontsize = 15)  #Labeling the x-axis
ylabel("Y-axis label", fontsize = 15)  #Labeling the y-axis
title("Title of the Graph", fontsize = 20)   #Giving the graph a Title
legend(loc = 4)  #Creating a legend which takes labels for each plot from when we plotted them and placing the legend in the bottom right corner
grid(True) #Creating a grid along the major tick marks
ax = gca()
ax.spines['right'].set_color('none') #Removing the borders of the graph
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
show()
```

###Subplots

```python
y = arange(-5,5, 0.01) #create a range in the x-axis over which to plot data
E = cosh(y)
p = sinh(y)

subplot(2, 1, 1)  #Specifying that the following data is plotted in the first position of a 2 row, 1 column set
title("Title of the Graph", fontsize = 20)  #The title
plot(y, p, color = "green", label = "Label1")  #The function to be plotted
legend(loc = 2)  #Establishing a legend in the top left corner
ylabel("Y-axis label", fontsize = 15)  #Labeling the y-axis
grid(True)  #Creating the grid

subplot(2, 1, 2)  #Specifying that the following data is plotted in the second position of the 2 row, 1 column set
plot(y, E, color = "red", label = "Label2") #The function to be plotted
legend(loc = 9) #Establishing a legend in the top center
xlabel("X-axis label", fontsize = 15)  #Labeling the common x-axis
ylabel("Y-axis label", fontsize = 15) #Labeling the y-axis
grid(True) #Creating the grid
show()
```

###Scatter Plots

```python
n = 200
x = random.normal(0,1,n)
y = random.normal(0,1,n)

scatter(x, y, color = "red", s=2)  #Plotting the data as a scatter plot with red dots with point2 size
xlabel("X-axis label", fontsize = 15)  #Labeling the x-axis
ylabel("Y-axis label", fontsize = 15)  #Labeling the y-axis
title("Title of the Graph", fontsize = 20)  #Giving the plot a title
legend()  #Creating the legend in the top right corner, the default position
grid(True)  #Creating the grid
ax = gca()
ax.spines['right'].set_color('none')  #Removing the borders of the graph
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
show()
```

```python
f = lambda x: x**(0.5) + sin(x)
```

```python
x = arange(0.0,15.0,0.001)
```

```python
myplot = plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin(x)$', fontsize=20)
xlabel(r'$x$', fontsize=20)
xlim(0.0,15.0)

show()
```

```python
xmin = 1.7
xmax = 12.4

#Find the maximum value of f within the boundary to sey ymax
subx = x[logical_and((x > xmin),(x < xmax))]
ymaz = max(f(subx))*1.05
ymin = 0

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.0,15.0)

#Plot the box
plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--') #vertical left
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--') #vertical right
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--') #horizontal top

show()
```

```python
#random_sample

N = 10000
samples_x = xmin + (xmax - xmin) * random_sample(N);
samples_y = ymin + (ymax - ymin) * random_sample(N);

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.0,15.0)

plot(samples_x[1:200],samples_y[1:200],'.')

#Plot the box
plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--') #vertical left
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--') #vertical right
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--') #horizontal top

show()
```

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

```python
#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.0,15.0)

plot(samples_x[1:200],samples_y[1:200],'.')

#Show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'.')

#Plot the box
plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--') #vertical left
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--') #vertical right
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--') #horizontal top

show()
```

```python
I = (xmax - xmin) * (ymax - ymin) * sum(newmask) / newmask.size
print I
```

```python
Iana = lambda x: (2.0/3.0) * x**(3.0/2.0) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = 100 * (Iexact - I)/Iexact

print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```

```python
f = lambda x: x**(0.5) + sin(x)
x = arange(0.0,15.0,0.001)

xmin = 1.7
xmax = 12.4

subx = x[logical_and((x > xmin),(x < xmax))]
ymaz = max(f(subx))*1.05
ymin = 0

N = 9000000
samples_x = xmin + (xmax - xmin) * random_sample(N);
samples_y = ymin + (ymax - ymin) * random_sample(N);

newmask = (samples_y < f(samples_x))

myplot = plot(x,f(x),lw=3)
xlim(0.0,15.0)

plot(samples_x[1:200000],samples_y[1:200000],'.')

plot(samples_x[newmask[:200000]],samples_y[newmask[:200000]],'.')

plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--') #vertical left
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--') #vertical right
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--') #horizontal top

show()
```

```python
I = (xmax - xmin) * (ymax - ymin) * sum(newmask) / newmask.size

Iana = lambda x: (2.0/3.0) * x**(3.0/2.0) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = 100 * (Iexact - I)/Iexact

print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```

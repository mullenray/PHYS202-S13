Electric Dipole
=

\begin{equation}
V(x, y)= \frac{kq}{\sqrt{x^2 + (y - d)^2}} - \frac{kq}{\sqrt{x^2 + (y + d)^2}}
\end{equation}

```python
dx = 0.06
dy = 0.06

x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = meshgrid(x,y)

k = 1; q = 1; d = 1
V = (k*q/(X**2 + (Y-d)**2)**(1/2.0)) - (k*q/(X**2 + (Y + d)**2)**(1/2.0))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.RdYlGn_r)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink=0.5)
```

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
contour(X,Y,V, levels, linewidths = 3, cmap = cm.Dark2)
show()
```

Ex. 2
=

\begin{equation}
Q(x, y)= \frac{kq}{\sqrt{x^2 + (y - d)^2}} + \frac{kq}{\sqrt{x^2 + (y + d)^2}} - \frac{kq}{\sqrt{(x - d)^2 + y^2}} - \frac{kq}{\sqrt{(x + d)^2 + y^2}}
\end{equation}

```python
da = 8.0/101.0
db = 8.0/101.0

a = np.arange(-4.0,4.0,da)
b = np.arange(-4.0,4.0,db)
A,B = meshgrid(a,b)

k = 8.987551787 #(Nm^2)/C^2
q = 1**(-9) #C
d = 2 #m
Q = (k*q/(A**2 + (B - d)**2)**(1/2.0)) + (k*q/(A**2 + (B + d)**2)**(1/2.0)) - (k*q/((A - d)**2 + B**2)**(1/2.0)) - (k*q/((A + d)**2 + B**2)**(1/2.0))

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
pl = ax.plot_surface(A, B, Q, rstride=1, cstride=1, linewidth=0, cmap=cm.RdYlGn_r)
cb = fig.colorbar(pl, shrink=0.5)
```

```python
levels = array([-160, -80, -40, -20, -10, -5, -2.5, -1.25, 0, 1.25, 2.5, 5, 10, 20, 40, 80, 160])
contour(A,B,Q, levels, linewidths = 1, cmap = cm.Dark2)
show()
```

```python
def pointPotential(x,y,q,posx,posy):
    """x,y are the point or points in space for which the electric potential is to be found.
    q is the charge
    posx, posy is the position of the charge in the plane

    returns the electric potential due to the charge"""
    k = 8.987551787 #(Nm^2)/C^2
    Vxy = ((k * q)/((x - posx)**2 + (y - posy)**2)**(1/2.0))
    return Vxy
```

```python
x = 1.0; y = 1.0
V11 = pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
def dipolePotential(x,y,q,d):
    """x,y are the point or points in space for which the electric potential is to be found.
    q is the charge
    d is the separation distance between the charges

    returns the electric potential due to the dipole"""
    V1 = pointPotential(x,y,q,abs(d),0.0)
    V2 = pointPotential(x,y,q,-abs(d),0.0)
    Vxy = V1 - V2
    return Vxy
```

```python
dx = 0.06
dy = 0.06

x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = meshgrid(x,y)

V = dipolePotential(X,Y,1e-9,1)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.RdYlGn_r)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink=0.5)
```

You are given the following information, but you may prefer to do some research for yourself.

*   1 Jan 1900 was a Monday.
*   Thirty days has September,
    April, June and November.
    All the rest have thirty-one,
    Saving February alone,
    Which has twenty-eight, rain or shine.
    And on leap years, twenty-nine.
*   A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?

```python
"""Setting the initial values that will be used by the code"""
ans = 0       #The total number of Sundays that fall on the first of a month 
dayweek = 1   #Keeping track of the day of the week: Sunday = 0, Monday = 1, etc.
year = 1      #Keeping track of the year starting at 1901

"""Defining functions to return the day of the week."""
def newdayweek30(dayweek):
    """For a month with 30 days, returns what day of the week the month ends on given the day of the week the previous month ends on."""
    if dayweek in range(0,5):
        dayweek += 2
        return dayweek
    else:
        dayweek -= 5
        return dayweek
        
def newdayweek31(dayweek):
    """For a month with 31 days, returns what day of the week the month ends on given the day of the week the previous month ends on."""
    if dayweek in range(0,4):
        dayweek += 3
        return dayweek
    else:
        dayweek -= 4
        return dayweek
        
def newdayweekFebL(dayweek):
    """For February during a leap year, returns what day of the week the month ends on given the day of the week the previous month ends on."""
    if dayweek in range(0,6):
        dayweek += 1
        return dayweek
    else:
        dayweek = 0
        return dayweek
        
def regyear(dayweek, ans):
    """For a regular year (not a leap year), returns the day of the week the year ends on given the day of the previous year ended on and returns a subtotal of the answer."""
    if dayweek == 6:
        ans += 1
    dayweek1 = newdayweek31(dayweek)
    if dayweek1 == 6:
        ans += 1
    dayweek2 = dayweek1
    if dayweek2 == 6:
        ans += 1
    dayweek3 = newdayweek31(dayweek2)
    if dayweek3 == 6:
        ans += 1
    dayweek4 = newdayweek30(dayweek3)
    if dayweek4 == 6:
        ans += 1
    dayweek5 = newdayweek31(dayweek4)
    if dayweek5 == 6:
        ans += 1
    dayweek6 = newdayweek30(dayweek5)
    if dayweek6 == 6:
        ans += 1
    dayweek7 = newdayweek31(dayweek6)
    if dayweek7 == 6:
        ans += 1
    dayweek8 = newdayweek31(dayweek7)
    if dayweek8 == 6:
        ans += 1
    dayweek9 = newdayweek30(dayweek8)
    if dayweek9 == 6:
        ans += 1
    dayweek10 = newdayweek31(dayweek9)
    if dayweek10 == 6:
        ans += 1
    dayweek11 = newdayweek30(dayweek10)
    if dayweek11 == 6:
        ans += 1
    dayweek12 = newdayweek31(dayweek11)
    return dayweek12, ans

def leapyear(dayweek, ans):
    """For a leap year, returns the day of the week the year ends on given the day of the previous year ended on and returns a subtotal of the answer."""
    if dayweek == 6:
        ans += 1
    dayweek1 = newdayweek31(dayweek)
    if dayweek1 == 6:
        ans += 1
    dayweek2 = newdayweekFebL(dayweek1)
    if dayweek2 == 6:
        ans += 1
    dayweek3 = newdayweek31(dayweek2)
    if dayweek3 == 6:
        ans += 1
    dayweek4 = newdayweek30(dayweek3)
    if dayweek4 == 6:
        ans += 1
    dayweek5 = newdayweek31(dayweek4)
    if dayweek5 == 6:
        ans += 1
    dayweek6 = newdayweek30(dayweek5)
    if dayweek6 == 6:
        ans += 1
    dayweek7 = newdayweek31(dayweek6)
    if dayweek7 == 6:
        ans += 1
    dayweek8 = newdayweek31(dayweek7)
    if dayweek8 == 6:
        ans += 1
    dayweek9 = newdayweek30(dayweek8)
    if dayweek9 == 6:
        ans += 1
    dayweek10 = newdayweek31(dayweek9)
    if dayweek10 == 6:
        ans += 1
    dayweek11 = newdayweek30(dayweek10)
    if dayweek11 == 6:
        ans += 1
    dayweek12 = newdayweek31(dayweek11)
    return dayweek12, ans

while year <= 100:     #For each year within the century,
    if year % 4 == 0:  #determines if it is a leap year,
        dayweek, ans = leapyear(dayweek, ans)  
        year += 1
    else:
        dayweek, ans = regyear(dayweek, ans)
        year += 1
        
print ans
```

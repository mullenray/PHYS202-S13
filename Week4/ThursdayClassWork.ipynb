Using Boolean Masks for Finding Data
=

```python
theta = arange(0., 2*pi, pi/100)
plot(theta, sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
#print mask
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
show()
```

```python
y = sin(theta[mask])
jj = (y == max(y))
y[jj] = NaN
plot(theta,sin(theta))
plot(theta[mask],y,linewidth=4)
show()
```

```python
thetaNew = arange(0,2*pi+0.1,pi/30)
sinNew = interp1d(theta, sin(theta))(thetaNew)
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
```

```python
plot(thetaNew,sinNew - sin(thetaNew))
ylabel("difference between interpolated and exact values")
show()
```

Differentiation
=

```python
x = arange(0.,10.01,0.1)
y = x**5
```

```python
plot(x,y)
show()
```

```python
dy = diff(y)  #y(n+1) - y(n)
print len(y), len(dy)
dx = diff(x)
dydx = dy/dx
print len(dydx)
```

```python
plot(x,dydx)
```

```python
dydx1 = np.zeros(y.shape,float)
dydx1[:-1] = diff(y)/diff(x)  #from start of array to one away from the end
dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])   #Manually input value of last element
```

$y = x^5$, $\frac{dy}{dx} = 5x^4$

```python
#calculate dydx by center differencing using array slices
dydx2 = zeros(y.shape,float)
dydx2[1:-1] = (y[2:] - y[:-2])/(x[2:]-x[:-2]) #center difference
dydx2[0] = (y[1]-y[0])/(x[1]-x[0])  #forward difference for first element
dydx2[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])  #backward diference for last element
```

```python
dydxExact = 5*x**4
```

```python
plot(x,dydx1, label='forward difference')
plot(x,dydx2, label='center difference')
plot(x,dydxExact, label='exact')
legend(loc='upper left')
show()
```

```python
percentError1 = 100*abs( dydx1[1:] - dydxExact[1:] )/dydxExact[1:]
percentError2 = 100*abs( dydx2[1:] - dydxExact[1:] )/dydxExact[1:]

#Use a semilog y-axis to see the variation better
semilogy(x[1:], percentError1, label='forward difference')
semilogy(x[1:], percentError2, label='center difference')
ylabel("percent error with dydxExact")
legend(loc="upper right")
show()
```

```python
from scipy.interpolate import interp1d
```

```python
def Temp(Resistance):
    """Input a Resistance (in units of mega Ohms) and output the temperature (in units of deg C) associated with that resistance."""
    temp = arange(10,40,1)
    res = [3.239,3.118,3.004,2.897,2.795,2.700,2.620,2.526,2.446,2.371,2.300,2.233,2.169,2.110,2.053,2.000,1.950,1.902,\
           1.857,1.815,1.774,1.736,1.700,1.666,1.634,1.603,1.574,1.547,1.521,1.496]
    resran = interp1d(res[::-1],temp)
    inttem = resran(Resistance)
    return inttem
```

###(b)

```python
print Temp(3.0)
print Temp(2.700)
print Temp(2.2)
print Temp(1.75)
```

###(c)

```python
R = arange(1.5,3.2,0.1)
T = Temp(R)

plot(R,T, color="green")
xlabel("$Resistance (M\Omega)$", fontsize = 15)
ylabel("$Temperature (^{\circ}C)$", fontsize = 15)
title("Temp(Resistance) vs. Resistance", fontsize = 18)
grid(True)
text(1.8,18,"$18.37^{\circ}C$ for $1.75 M\Omega$",fontsize=15)
show()
```

###(a)

```python
def finiteDifference(x,y):
    """Input array x and y(x).
       Return the derivative of y with respect to x using the center difference method."""
    dydx = np.zeros(y.shape, float)
    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:]-x[:-2]) #center difference
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])  #forward difference for first element
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])  #backward diference for last element
    return dydx
    
```

###(b)

```python
x = arange(0.0,10.01,0.1)
y = x**5

dydx1 = finiteDifference(x,y)
print dydx1
```

###(c)

```python
z = arange(0.0,10.0005,0.001)
w = z**5
dydx2 = finiteDifference(z,w)
```

```python
dydxExact1 = 5*x**4
dydxExact2 = 5*z**4

percentError1 = 100*abs( dydx1[1:] - dydxExact1[1:] )/dydxExact1[1:]
percentError2 = 100*abs( dydx2[1:] - dydxExact2[1:] )/dydxExact2[1:]

semilogy(x[1:], percentError1, label='Steps of 0.1', color = "green", linewidth = 2)
semilogy(z[1:], percentError2, label='Steps of 0.001', color = "red", linewidth = 2)
ylabel("Percent error with dydxExact", fontsize = 13)
legend()
show()
```

###(b)

```python
def fourPtFiniteDiff(x,y):
    """Input arrays x and y(x).
       Computes the numerical derrivative of y with respect to x using the four-point centre differencing method."""
    dydx = np.zeros(y.shape,float)
    for i,n in enumerate(y[2:-2]):
        i += 2
        dydx[i] = (y[i - 2] - (8 * y[i - 1]) + (8 * y[i + 1]) - y[i + 2])/(12 * (x[i + 1] - x[i]))
    dydx[0] = (y[1]-y[0])/(x[1]-x[0]) #forward difference
    dydx[1] = (y[2]-y[1])/(x[2]-x[1]) #forward difference
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2]) #backward diference
    dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3]) #backward diference
    return dydx
```

```python
x = arange(0.0,10.01,0.1)
y = x**5

dydx = fourPtFiniteDiff(x,y)

dydxExact = 5*x**4
percentError = 100*abs( dydx[1:] - dydxExact[1:] )/dydxExact[1:]

semilogy(x[1:], percentError, color = "green", linewidth = 2)
grid(True)
title("Percent Error using Four-Point Center Differencing Method")
ylabel("Percent error with dydxExact", fontsize = 13)
show()
```

###(c)

```python
plot(x,y, label = "Original Data")
plot(x,finiteDifference(x,y), label = "2pt-Forward Diff", color = 'red')
plot(x,cos(x), label = 'Analytical Derivative', color = 'green', linewidth = 2)
plot(x,fourPtFiniteDiff(x,y), label = '4pt-Centered Diff', color = 'black', linestyle = "dashed", linewidth = 2)
legend(loc = 'upper center', prop = {'size' : 11})
ylabel("Noisy data and derivatives", size = 14)
xlabel("x", size = 14)
show()
```
